### 1.pandas 的索引
索引是非常重要的一个功能，使用索引我们就可以对数据进行选取和筛选

**四种索引形式：**
1. 使用位置做索引
2. 使用列表做索引
3. 使用切片做索引
4. 使用bool类型索引

** A、选取行: **

使用 loc 或者 iloc 函数来实现

```python
import pandas as pd
```

```python
filepath = '销售客户信息.xlsx'

df = pd.read_excel(filepath)
```

```python
df
```

```python
# 直接索引，获取第2 行
df.loc[2]
```

```python
# 2.列表索引，列表可以是arange，range，ndarray ，series 都可以，列表形式即可
# 获取0，1，2 行
df.loc[[0,1,2]]
```

```python
# 3.切片索引
df.loc[0:4]
```

```python
# 4.bool类型索引
b1 = df['年龄']>30
b1
```

```python
df.loc[b1]
```

**iloc 和 loc 的区别**
1. loc 是基于索引值的，切片是左闭右闭的
2. iloc 是基于位置的，切片是左闭右开的

```python
df.loc[0:2]
```

```python
df.iloc[2]
```

```python
df1 = df.set_index('用户ID')
df1
```

```python
df1.loc[0:2]
```

```python
df1.iloc[0:2]
```

```python
df.loc[0:5,'年龄']
```

```python
df1.iloc[0:5]
```

** B、选取列: **

不需要使用函数

```python
df['年龄'].head()
```

```python
df[['年龄','城市']].head()
```

```python
# 注意使用切片和bool索引的话，效果和选取行是一样的
df[0:3]
```

```python
df.iloc[0:3]
```

```python
df[df['年龄']>18]
```

```python
df.loc[df['年龄']>18]
```

** C、选取行列: **

使用切片

```python
df.loc[0,'年龄']
```

```python
df.loc[[0,3],['年龄','性别']]
```

```python
df.loc[0:3,['性别','年龄']]
```

```python
df.iloc[0:3,0:2]
```

```python
df.loc[df['年龄']>18,:]
```
